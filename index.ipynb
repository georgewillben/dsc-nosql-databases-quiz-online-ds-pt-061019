## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

1) MongoDB works by storing collections of documents. Each document has its own key-value pairs, can store different amounts of data, and each document does not have to share keys with other documents.
<br>
2) It is diffrent from a relational database because documents do not have to share the same keys. In a relational database every entry in the database within a table has to have the same fields (columns).






## Question 4

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
1) Fault tolerance is when data from one server is backed up on another server. This protects the dataset incase a server isn't running.<br>
2) This has to do with RDD's because they store data on many diffrent servers.
